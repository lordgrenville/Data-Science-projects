```python
def _eval_dfs_up_to_decimal(x, y):
    try:
        pd.testing.assert_frame_equal(
            x, y, check_exact=False,
            check_dtype=True,
            check_less_precise=int(decimals))
        equal = True
    except AssertionError:
        equal = False
    return equal
```

```python
def _f(x):
    s = str(x)
    if (len(s) == 19) and (s[10:] == ' 00:00:00'):
        return s[:10]
    return x
```

```python
def _smart_fill_na(df):
    d = {
        'object': '',
        'int64': -12345,
        'float64': -12345.0,
    }

    cols = df.columns.tolist()
    for col in cols:
        df[col] = df[col].fillna(d[str(df[col].dtype)])
    return df
```

```python
def _is_probably_datetime_column(df1, df2, col):
    if any([x in col.lower() for x in ['date', 'time']]):
        return True
    for i in range(10):
        s1 = df1.sample(10)
        if not (pd.isnull(s1.apply(lambda x: pd.to_datetime(str(x), errors='coerce'))).all()):
            return True
        s2 = df2.sample(10)
        if not (pd.isnull(s2.apply(lambda x: pd.to_datetime(str(x), errors='coerce'))).all()):
            return True
    return False
```

```python
dir_1 = 'old/'
dir_2 = 'new/'
filename_list_1 = ['account.csv','support.csv','target.csv','usage.csv',]
filename_list_2 = ['account.csv','support.csv','target.csv','usage.csv',]
decimals = 4
file_name_to_rename_dict = {}
```

```python
for file1, file2 in zip(filename_list_1, filename_list_2):

    full_path_1 = dir_1 + file1
    full_path_2 = dir_2 + file2
    identical_flag = filecmp.cmp(full_path_1, full_path_2)
    print('===')
    print("{} : {} <> {}".format('identical' if identical_flag else 'NOT indentical', file1, file2))
    if not identical_flag:
        df1 = pd.read_csv(full_path_1, dtype=object)
        df2 = pd.read_csv(full_path_2, dtype=object)

        if ('pai_account_id' in list(df1))and('date' in list(df1)):
            df1 = df1.sort_values(by=['pai_account_id', 'date']).reset_index(drop=True)
        if ('pai_account_id' in list(df2)) and ('date' in list(df2)):
            df2 = df2.sort_values(by=['pai_account_id', 'date']).reset_index(drop=True)

        if _eval_dfs_up_to_decimal(df1, df2):
            print("{} : {} <> {}".format('almost identical', file1, file2))
            continue

        cols = sorted(df1.columns.tolist())
        if cols != sorted(df2.columns.tolist()):
            s1 = set(df1.columns.tolist())
            s2 = set(df2.columns.tolist())
            s = 'ERROR columns: f1-f2 = {}\n'.format(sorted(list(s1.difference(s2))))
            s += 'ERROR columns: f2-f1 = {}\n'.format(sorted(list(s2.difference(s1))))
            print(s)
            if file1 in file_name_to_rename_dict:
                print('renaming {} columns {}'.format(file1, file_name_to_rename_dict[file1]))
                df1 = df1.rename(columns=file_name_to_rename_dict[file1])
            s1 = set(df1.columns.tolist())
            cols = list(s1.intersection(s2))
            df1 = df1[cols].copy()
            df2 = df2[cols].copy()

        if df1.shape != df2.shape:
            print('Shapes are NOT identical: {}!={}'.format(df1.shape, df2.shape))
            if not('pai_account_id' in list(df1) and 'pai_account_id' in list(df2) and 'date' in list(df1) and 'date' in list(df2)):
                continue
            s1 = set(zip(df1['pai_account_id'].tolist(), df1['date'].tolist()))
            s2 = set(zip(df2['pai_account_id'].tolist(), df2['date'].tolist()))
            s = s1.intersection(s2)
            print('number of common rows: {}'.format(len(s)))
            r = df1.apply(lambda row: (row['pai_account_id'], row['date']) in s, axis=1)
            df1 = df1.loc[r].reset_index(drop=True)
            r = df2.apply(lambda row: (row['pai_account_id'], row['date']) in s, axis=1)
            df2 = df2.loc[r].reset_index(drop=True)

        if df1.shape == df2.shape:
            print('Shapes are identical: {}=={}'.format(df1.shape, df2.shape))

            cols = df1.columns.tolist()
            if 'pai_account_id' in list(df1) and 'date' in list(df1):
                cols.remove('pai_account_id')
                cols.remove('date')
                cols = ['pai_account_id', 'date'] + cols
                df1 = df1.sort_values(by=cols).reset_index(drop=True)
                df2 = df2.sort_values(by=cols).reset_index(drop=True)

            for col in cols:
                if _eval_dfs_up_to_decimal(df1[[col]], df2[[col]]):
                    print('{} is indentical'.format(col))
                    continue

                r = pd.isnull(df1[col]) | pd.isnull(df2[col])
                x = pd.to_numeric(df1.loc[~r][col], errors='coerce').to_frame()
                y = pd.to_numeric(df2.loc[~r][col], errors='coerce').to_frame()
                if (~pd.isnull(x[col])).all() and (~pd.isnull(y[col])).all() and x.shape[0]>0:
                    x = x[[col]].values.astype(float)
                    y = y[[col]].values.astype(float)
                    if (x == y).all():
                        print('{} is indentical'.format(col))
                        continue
                else:
                    if _is_probably_datetime_column(df1[col], df2[col], col):
                        df1[col] = pd.to_datetime(df1[col], errors='coerce')
                        df2[col] = pd.to_datetime(df2[col], errors='coerce')

                if _eval_dfs_up_to_decimal(df1[[col]], df2[[col]]):
                    print('{} is indentical'.format(col))
                    continue

                rows = df1[col] != df2[col]
                rows2 = pd.isnull(df1[col])
                rows2 = np.logical_and(rows2, pd.isnull(df2[col]))
                rows2 = np.logical_not(rows2)
                rows = np.logical_and(rows, rows2)
                print('ERROR {}:{} mismatches out of {}'.format(col, rows.sum(), rows.size))

                l1 = df1.loc[rows][col].tolist()
                l2 = df2.loc[rows][col].tolist()

                if not ('pai_account_id' in list(df1) and 'pai_account_id' in list(df2) and 'date' in list(
                        df1) and 'date' in list(df2)):
                    print(list(zip(l1,l2))[:10])

                ddf1 = df1.loc[rows][['pai_account_id', 'date', col]].copy()
                ddf2 = df2.loc[rows][['pai_account_id', 'date', col]].copy()

                print(ddf1.head())
                print(ddf2.head())
```

```python
def reader(file, **kwargs):
    """
    wrapper for read_csv that tries a bunch of different encodings before giving up
    """
    encodings = ['utf-8', 'iso-8859-1', 'cp1252', 'latin1']
    for encoding in encodings:
        try:
            return pd.read_csv(file, encoding=encoding, **kwargs)
        except UnicodeDecodeError as e:
            continue
    return e


def shared_val(df1, df2, col):
    a = set(df1[col].values)
    b = set(df2[col].values)
    return list(a & b)  # intersection


def column_diff(df1, df2):
    one_cols, two_cols = set(list(df1)), set(list(df2))
    left = one_cols - two_cols
    right = two_cols - one_cols
    return [left, right]


def cleaning(mine, si, renaming_dict=None):
    # mine.client_id = 5
    if rename_dict:
        mine.rename(columns=rename_dict, inplace=True)
    mine = mine[list(si)]
    mine, si = [df.sort_values(by=['pai_account_id', 'date']).reset_index(drop=True)
                for df in [mine, si]]
    mine.replace(np.nan, 'nan', inplace=True)
    si.replace(np.nan, 'nan', inplace=True)
    return mine, si


def get_joint(mine, si):
    shared_dates = shared_val(mine, si, 'date')
    shared_acc = shared_val(mine, si, 'pai_account_id')

    mine = mine[mine.date.isin(shared_dates)]
    si = si[si.date.isin(shared_dates)]
    mine = mine[mine.pai_account_id.isin(shared_acc)]
    si = si[si.pai_account_id.isin(shared_acc)]
    return mine, si


def move_column_to_front(df, cols_to_move):
    cols = list(df)
    for col in cols_to_move:
        cols.insert(0, cols.pop(cols.index(col)))
    return df.reindex(columns=cols)


def value_diff(df1, df2, col_list):
    res = []
    for col in col_list:
        print('Items in df1 **{}** but not in df2:'.format(col))
        diff_l = set(df1[col].unique()) - set(df2[col].unique())
        print(len(diff_l))
        print('Items in df2 **{}** but not in df1:'.format(col))
        diff_r = set(df2[col].unique()) - set(df1[col].unique())
        print(len(diff_r))
        res.extend([diff_l, diff_r])
    return res


def checker(mine, si, rename_dict=None, verbose=True):
    all_good =True
    sh1, sh2 = mine.shape, si.shape
    if sh1 != sh2:
        return('Shapes unequal: 1:{}, 2:{}'.format(sh1,sh2))
    else:
        print('Shapes equal. Checking rows - might take a while...')
#     mine, si = cleaning(mine, si, renaming_dict=rename_dict)
    print('column_diff: {}'. format(column_diff(mine, si)))
    cols = list(zip(list(mine), list(si)))
    issues = []
    for a in range(len(mine)):
        row1, row2 = mine.iloc[a].values, si.iloc[a].values
        if not np.array_equal(row1, row2):
            if all_good:
                print('Error found...')
                all_good=False
            if verbose:
                print('Inequality in row # {}'.format(a))
            issues.append(a)
            for idx, pair in enumerate(list(zip(row1, row2))):
                if pair[0] != pair[1] and verbose:
                    print('In col {}'.format(cols[idx]))
                    print(pair[0], pair[1])
            print(row1, row2)
            break  # comment out to return all
    if all_good:
        return'Woohoo! All good'
    else:
        return issues
```
