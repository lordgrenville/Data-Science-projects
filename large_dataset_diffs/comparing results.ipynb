```python
def my_checker(t1, t2):
    t1 = t1[sorted(list(t1))]
    t2 = t2[sorted(list(set(list(t2)) & set(list(t1))))]
    joint = t1.join(t2, how='inner', lsuffix='_old', rsuffix='_new')
    assert joint.shape[0] == t1.shape[0] == t2.shape[0]
    n = int(joint.shape[1] / 2)
    col_names, bad_cols = list(joint), list()
    for (i,j) in (zip(range(n), range(n, 2 * n))):
        try:
            assert_series_equal(joint.iloc[:,i], joint.iloc[:,j], check_names=False, check_less_precise=True)
        except AssertionError as e:
            print('Prob in col {}, {}!'.format(col_names[i], col_names[j]), str(e)[:62])
            bad_cols.append(col_names[i][:-4])
    return joint, bad_cols
```

```python
def _eval_dfs_up_to_decimal(x, y):
    try:
        pd.testing.assert_frame_equal(
            x, y, check_exact=False,
            check_dtype=True,
            check_less_precise=int(decimals))
        equal = True
    except AssertionError:
        equal = False
    return equal
```

```python
def _f(x):
    s = str(x)
    if (len(s) == 19) and (s[10:] == ' 00:00:00'):
        return s[:10]
    return x
```

```python
def _smart_fill_na(df):
    d = {
        'object': '',
        'int64': -12345,
        'float64': -12345.0,
    }

    cols = df.columns.tolist()
    for col in cols:
        df[col] = df[col].fillna(d[str(df[col].dtype)])
    return df
```

```python
def _is_probably_datetime_column(df1, df2, col):
    if any([x in col.lower() for x in ['date', 'time']]):
        return True
    for i in range(10):
        s1 = df1.sample(10)
        if not (pd.isnull(s1.apply(lambda x: pd.to_datetime(str(x), errors='coerce'))).all()):
            return True
        s2 = df2.sample(10)
        if not (pd.isnull(s2.apply(lambda x: pd.to_datetime(str(x), errors='coerce'))).all()):
            return True
    return False
```
