```python
def table_analyser(df):
    data = []
    for col in list(df):
        data.append([float((df[col].isna().sum() / df[col].shape[0]) * 100), df[col].nunique()])
    return pd.DataFrame(data, index=list(df), columns=['NaNs', 'Uniques'])#, dtype={'NaNs':int})
```
