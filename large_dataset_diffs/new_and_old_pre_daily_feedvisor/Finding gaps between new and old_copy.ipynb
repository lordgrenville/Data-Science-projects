```python
def check_shape(df1, df2):
    print(df1.shape, df2.shape)

    a = set(list(df1)) - set(list(df2))
    b = set(list(df2)) - set(list(df1))
    c = set(df1.index.get_level_values('date')) - set(df2.index.get_level_values('date'))
    d = set(df2.index.get_level_values('date')) - set(df1.index.get_level_values('date'))
    e = set(df1.index.get_level_values('pai_account_id')) - set(df2.index.get_level_values('pai_account_id'))
    f = set(df1.index.get_level_values('pai_account_id')) - set(df2.index.get_level_values('pai_account_id'))
    g = set(df1.index) - set(df2.index)
    h = set(df2.index) - set(df1.index)
    for s in zip([a, b, c, d, e, f, g, h], ['Cols in df1 but not df2', 
                                           'Cols in df2 but not df1',
                                           'Dates in df1 but not df2',
                                           'Dates in df2 but not df1',
                                           'PIDs in df1 but not df2',
                                           'PIDs in df2 but not df1',
                                           'Indices in df1 but not df2',
                                           'Indices in df2 but not df1',]):
        if len(s[0]) != 0:
            print(s[1], s[0])
    return
```

```python
def my_checker(t1, t2):
    t1 = normalizer(t1)
    t2 = t2 = normalizer(t2)
    t1 = t1[sorted(list(t1))]
    t2 = t2[sorted(list(set(list(t2)) & set(list(t1))))]
    joint = t1.join(t2, how='inner', lsuffix='_old', rsuffix='_new')
    assert joint.shape[0] == t1.shape[0] == t2.shape[0]
    n = int(joint.shape[1] / 2)
    col_names = list(joint)
    for (i,j) in (zip(range(n), range(n, 2 * n))):
        try:
            assert_series_equal(joint.iloc[:,i], joint.iloc[:,j], check_names=False, check_less_precise=True)
        except AssertionError as e:
            print('Prob in col {}, {}!'.format(col_names[i], col_names[j]), str(e)[:62])
    return joint
```

```python
def _eval_dfs_up_to_decimal(x, y):
    try:
        pd.testing.assert_frame_equal(
            x, y, check_exact=False,
            check_dtype=True,
            check_less_precise=int(decimals))
        equal = True
    except AssertionError:
        equal = False
    return equal
```

```python
def _smart_fill_na(df):
    d = {
        'object': '',
        'int64': -12345,
        'float64': -12345.0,
    }

    cols = df.columns.tolist()
    for col in cols:
        df[col] = df[col].fillna(d[str(df[col].dtype)])
    return df
```

```python
def _is_probably_datetime_column(df1, df2, col):
    if any([x in col.lower() for x in ['date', 'time']]):
        return True
    for i in range(10):
        s1 = df1.sample(10)
        if not (pd.isnull(s1.apply(lambda x: pd.to_datetime(str(x), errors='coerce'))).all()):
            return True
        s2 = df2.sample(10)
        if not (pd.isnull(s2.apply(lambda x: pd.to_datetime(str(x), errors='coerce'))).all()):
            return True
    return False
```
